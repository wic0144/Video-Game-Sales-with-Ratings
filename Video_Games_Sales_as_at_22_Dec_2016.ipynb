```
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/wic0144/Video-Game-Sales-with-Ratings/main/Video_Games_Sales_as_at_22_Dec_2016.csv')
df.head()
```

```
NA_Sales = df[['Name','Genre','NA_Sales','Platform']]
EU_Sales = df[['Name','Genre','EU_Sales','Platform']]
JP_Sales = df[['Name','Genre','JP_Sales','Platform']]
Global_Sales = df[['Name','Genre','Global_Sales','Platform']]
```

```

Max_NA = NA_Sales.loc[df['NA_Sales'].idxmax()]
Max_EU = EU_Sales.loc[df['EU_Sales'].idxmax()]
Max_JP = JP_Sales.loc[df['JP_Sales'].idxmax()]
Max_Global = Global_Sales.loc[df['Global_Sales'].idxmax()]
list=(Max_NA,Max_EU,Max_JP,Max_Global)
```

```
Category_NA_Sales=NA_Sales.drop(columns=(['Name','Platform'])).groupby(by='Genre').sum().sort_values('NA_Sales',ascending=False)
Category_EU_Sales=EU_Sales.drop(columns=(['Name','Platform'])).groupby(by='Genre').sum().sort_values('EU_Sales',ascending=False)
Category_EU_Sales.head()
Category_NA_Sales.head()
```

```
TOP20_NA = NA_Sales.sort_values('NA_Sales',ascending=False).iloc[0:20].drop(columns='Genre')
TOP20_EU = EU_Sales.sort_values('EU_Sales',ascending=False).iloc[0:20].drop(columns='Genre')
TOP20_JP = JP_Sales.sort_values('JP_Sales',ascending=False).iloc[0:20].drop(columns='Genre')
TOP20_Gobal = Global_Sales.sort_values('Global_Sales',ascending=False).iloc[0:20].drop(columns='Genre')
```

```
TOP20_NA = NA_Sales.sort_values('NA_Sales',ascending=False).iloc[0:20].drop(columns='Genre')
TOP20_EU = EU_Sales.sort_values('EU_Sales',ascending=False).iloc[0:20].drop(columns='Genre')
TOP20_JP = JP_Sales.sort_values('JP_Sales',ascending=False).iloc[0:20].drop(columns='Genre')
TOP20_Gobal = Global_Sales.sort_values('Global_Sales',ascending=False).iloc[0:20].drop(columns='Genre')

fig, axes = plt.subplots(nrows=2, ncols=2)

plt.subplots_adjust(wspace=1, hspace=0.5)

TOP20_NA.plot(ax=axes[0, 0],kind='barh',x='Name',figsize=(20,20),color='r');

axes[0, 0].set_title('NA_Sales');

TOP20_EU.plot(ax=axes[0, 1],kind='barh',x='Name',figsize=(20,20),color='g');

axes[0, 1].set_title('EU_Sales');

TOP20_JP.plot(ax=axes[1, 0],kind='barh',x='Name',figsize=(20,20),color='b');

axes[1, 0].set_title('JP_Sales');

TOP20_Gobal.plot(ax=axes[1, 1],kind='barh',x='Name',figsize=(20,20),color='y');

axes[1, 1].set_title('Gobal_Sales');
```
